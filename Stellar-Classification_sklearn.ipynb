```python
import pandas as pd
import time
```

```python
train = pd.read_csv('train.csv')
train
```

```python
test = pd.read_csv('test.csv')
test
```

```python
X_train = train[['u', 'g', 'r', 'i', 'z', 'redshift']]
#X_train = train[['alpha', 'delta', 'u', 'g', 'r', 'i', 'z', 'redshift']]
X_train
```

```python
y_train = train['label']
y_train
```

```python
X_test = test[['u', 'g', 'r', 'i', 'z', 'redshift']]
#X_test = test[['alpha', 'delta', 'u', 'g', 'r', 'i', 'z', 'redshift']]
X_test
```

```python
from sklearn.neural_network import MLPClassifier
```

```python
start_time = time.time()
clf_adam = MLPClassifier(solver='adam', random_state=9527, max_iter=500).fit(X_train, y_train)
end_time = time.time()
print(f'time: {end_time - start_time}s')
```

```python
start_time = time.time()
clf_sgd = MLPClassifier(solver='sgd', random_state=9527, max_iter=500).fit(X_train, y_train)
end_time = time.time()
print(f'time: {end_time - start_time}s')
```

```python
predict_adam = pd.DataFrame({'id': test.id})
predict_adam['label'] = clf_adam.predict(X_test)
predict_adam
```

```python
predict_sgd = pd.DataFrame({'id': test.id})
predict_sgd['label'] = clf_sgd.predict(X_test)
predict_sgd
```

```python
predict_adam.to_csv('submission_adam.csv', index=False)
predict_sgd.to_csv('submission_sgd.csv', index=False)
```
